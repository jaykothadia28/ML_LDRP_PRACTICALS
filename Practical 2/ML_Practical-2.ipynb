```python
import numpy as np

import pandas as pd
 
import matplotlib.pyplot as plt
 
import statsmodels.api as sm
 
import seaborn
 
seaborn.set()
```

```python
df = pd.read_csv('Real Estate.csv')
df.head()
```

```python
x1 = df[['X2 house age']]
y = df[['Y house price of unit area']]
```

```python
x = sm.add_constant(x1)
results_x1 = sm.OLS(y, x).fit()
results_x1.summary()
```

```python
x2 = df[['X3 distance to the nearest MRT station']]
x = sm.add_constant(x2)
results_x2 = sm.OLS(y, x).fit()
results_x2.summary()
```

```python
x3 = df[['X4 number of convenience stores']]
x = sm.add_constant(x3)
results_x3 = sm.OLS(y, x).fit()
results_x3.summary()
```

```python
x4 = df[['X5 latitude', 'X6 longitude']]
x = sm.add_constant(x4)
results_x4 = sm.OLS(y, x).fit()
results_x4.summary()
```

```python
x_final = df[['X2 house age', 'X3 distance to the nearest MRT station', 'X4 number of convenience stores']]
#print(x_final.head())
x = sm.add_constant(x_final)
results_final = sm.OLS(y, x).fit()
results_final.summary()
```
