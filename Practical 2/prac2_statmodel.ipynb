```python
import numpy as np

import pandas as pd
 
import matplotlib.pyplot as plt
 
import statsmodels.api as sm
 
import seaborn
 
seaborn.set()
```

```python
df = pd.read_csv('insurance.csv')
df.head()
```

```python
x1= df[['age']]

y=df[['charges']]
```

```python
x = sm.add_constant(x1)
 
results = sm.OLS(y,x).fit()
```

```python
results.summary()
```
